```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bn_dataframe = pd.read_csv('../input/bnda-data/history.csv')
da_dataframe = pd.read_csv('../input/da-data/history.csv')
bnda_dataframe = pd.read_csv('../input/bn-data/history.csv')
model4_dataframe = pd.read_csv('../input/model4-data/history4.csv')
```

```python
# Plot the loss vs. epoch curves for each method
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
histories = [bn_dataframe, da_dataframe, bnda_dataframe, model4_dataframe]
plt.figure(figsize=(10, 6))
for i, history in enumerate(histories):
    plt.plot(history['loss'], label=f'Method {i+1}', color=colors[i])
    plt.plot(history['val_loss'], label=f'Method {i+1}', color=colors[i], marker='o', markersize=4, linestyle='--')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs. Epochs for Different Methods')
plt.legend()
plt.savefig('Loss vs. Epochs.png')
plt.show()
```

```python
# Plot the accuracy vs. epoch curves for each method
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
plt.figure(figsize=(10, 6))
for i, history in enumerate(histories):
    plt.plot(history['accuracy'], label=f'Method {i+1}', color=colors[i])
    plt.plot(history['val_accuracy'], label=f'Method {i+1}', color=colors[i], marker='o', markersize=4, linestyle='--')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Epochs for Different Methods')
plt.legend()
plt.savefig('Accuracy vs. Epochs.png')
plt.show()
```
